Definamos la función y su segunda derivada

```julia
function f(x)
    return sin(6*(x^2)+1)
end
```

```julia
function segderAn(x)
    return 12*cos(6*(x^2)+1)-12*(x^2)*sin(6*(x^2)+1)
end
```

como nos piden el error absoluto en la segunda derivada de la función, busco una expresión para la segunda derivada $f''(a)$, obteniendo $$f''(a) \approx \frac{f(a+h) + f(a-h) -2f(a)}{h^2}$$

```julia
function segder(f,a,h)
    return (f(a+h)+f(a-h)-2*f(a))/(h^2)
end
```

```julia
#De forma numerica para el punto fijo 2 en el intervalo [1,3] y para dos valores distintos de h
println(segder(f,2,0.001))
println(segderAn(2))
# ver el error absoluto
println(abs(segderAn(2)-segder(f,2,0.001)))
```

```julia
println(segder(f,2,0.01))
println(segderAn(2))
# ver el error absoluto
println(abs(segderAn(2)-segder(f,2,0.01)))
```

De lo anterior podemos notar una variacion del valor absoluto, se realiza una grafica del error del valor absoluto entre la derivada en el punto fijo perteneciente al intervalo $[1,3]$ y su aproximación para distintos valores de $h$ entre 0 y 0.1, ya que nos interesan valores pequeños de h.

```julia
xs=range(0,stop=0.1,length=50)
x=2
ys=[abs(segderAn(x)-segder(f,x,h)) for h in xs]
plot(xs,ys, xlabel="x", ylabel="y")
```

Podemos observar en la gráfica que entre h más pequeña el error absoluto será mas grande.

## 4. Ejercicio 9. Clase 6
Sea $g:\mathbb{R}^2 \to \mathbb{R}$, obtén una expresión **analítica** (i.e. no deben programar nada) de diferencias finitas para la derivada curzada $\frac{\partial^2 g}{\partial x \partial y}$. ¿Cuál es el orden de su error de aproximación?

**Sugerencia:** Hay dos maneras de obtener la expresión, y para ambas debes suponer que las parciales cruzadas son iguales $\frac{\partial^2 f}{\partial x\partial y} = \frac{\partial^2 f}{\partial y \partial x}$. La manera larga es usar una serie de Taylor Multidimensional y resolver el sistema de ecuaciones que se obtiene. La manera corta es realizar dos aproximaciones simultáneas en cada entrada. 

Tenermos para la derivada parcial multidimensional 
$$
\frac{\partial f}{\partial x_i} = \lim_{h\to 0 }  \frac{f(a_1, \ldots, a_i+h,\ldots,a_n) - f(a_1, \ldots, a_i,\ldots,a_n)}{h}
$$
en este caso como $g:\mathbb{R}^2 \to \mathbb{R}$ tenemos las siguinetes ecuaciones
$$
\frac{\partial g}{\partial x} (x,y) = \lim_{h\to 0 }  \frac{g(x+h,y) - g(x,y)}{h}
$$
$$
\frac{\partial g}{\partial y} (x,y) = \lim_{h\to 0 }   \frac{g(x,y+h) - g(x,y)}{h}
$$

 obtenemos para la parcial cruzada
 $$
\frac{\partial ^2 g}{\partial x \partial y} = \lim_{h\to 0 } (\lim_{h\to 0 } \frac{g(x,y)- g(x+h,y)-g(x,y+h)+g(x+h,y+h)}{h^2}
$$

```julia
#Tomando la funcion x^2 para poder probar la expresión 
function g(x,y)
    return (x^2)*y
end
```

```julia
#sabemos que la derivada de la funcion es g(x)'=2x
function deranalit(x,y)
    return 2x
end
```

```julia
function parcruzada(g,x,y,h)
    a= g(x,y)- g(x+h,y)-g(x,y+h)+g(x+h,y+h)
    b=h^2
    return a/b
end
```

```julia
println(parcruzada(g,1,1,0.001))
println(deranalit(1,1))
# ver el error absoluto
print("Error:")
println(abs(deranalit(1,1)-parcruzada(g,1,1,0.001)))
```

Fija $n = 100$ y utiliza la función `miHeron(n,s)` con distintos valores de $s>0$. Analizando el arreglo obtenido, ¿Puedes ver si converge la sucesión? ¿De ser así, el número al que converge tiene alguna relación con $s$?

**Sugerencia:** puedes graficar la sucesión para analizar su convergencia.

```julia
function miHeron(n,s)
    #creo el arreglo elemento a elemento, la sucesión recursiva es para valores K mayores o iguales a uno, considerando esto mi primer elemento será 1, agrego el punto para cusar flotantes
   element=[1.0]
    # creamos todos los subsecuentes
    for k in 2:n
        push!(element,(0.5)*(element[k-1]+(s/(element[k-1]))))
    end
    # regreso el arreglo
    return element
end
```

```julia
#Fijo n=100 y ocupo s=80
miHeron(100,80)
```

la sucesión converge a $\sqrt s $ , realizo las siguientes dos graficas

```julia
function convergenci(s)
    xs = range(0,stop=50,length=100)
    ys = [miHeron(100,s) for s in xs]
    plot(xs,ys)
end
```

```julia
convergenci(100)
```

```julia
function convergenc(s)
    #como converge a sqrt(s) busco estar una unidad mas adelante para cada s en mi stop
    xs = range(0,stop=(s^(1/2))+1,length=100)
    ys = [ s^(1/2) for x in xs]
    yss= miHeron(100,s)
    plot(xs,ys,label="s^(1/2)")
    plot!(xs,yss,label="miHeron")
end
```

```julia
function f1(x)
    return (-(2*x)/(7))+3
end
```

```julia
function f2(x)
    return cos(x)-x
end
```

```julia
function mibiseccion(f,a,b,n)
intervalomin=a
intervalomax=b
    pm = (intervalomin+intervalomax)/2
    final=[pm]
    for i in 1:n
        if f(intervalomin)*f(pm) <= 0
            intervalomax = pm
        elseif f(intervalomax)*f(pm) <= 0
            intervalomin = pm
        end
        pm = (intervalomin + intervalomax)/2
        push!(final,pm)
    end
    return final
end
```

```julia
mibiseccion(f2,0,2,10)
```
