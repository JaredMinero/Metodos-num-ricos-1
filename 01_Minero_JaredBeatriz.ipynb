**Sugerencia:** ¿Qué representa la magnitud del producto cruz entre dos vectores?

**R** Recordemos el producto vectorial. Sean a y b vectores, definimos el producto vectorial como
$$a x b=
\left|
\begin{array}{cc}
i & j & k\\
a_{1} & a_{2} & a_{3}\\
b_{1} & b_{2} & b_{3}
\end{array}
\right|= i(a_{2}b_{3}-b_{2}a_{3})-j(a_{1}b_{3}-b_{1}a_{3})+k(a_{1}b_{2}-b_{1}a_{2})$$

```julia
function lados(Lx,Ly,Lz)
    x=[Lx,0,0]
    y=[0,Ly,0]
    z=[0,0,Lz]
    a=y+z
    b=x-y
    #realizando el producto vectorial entre a y b, obtenemos a x b=[y*z, -x*z, x*y]
    #obtenemos la norma 
    norma=sqrt((Ly*Lz)^2 + (Lx*Lz)^2 + (Lx*Ly)^2)
    area= norma / 2
    return area
end 
```

# 6. Ejercicio 2 clase 5.2
Compara las velocidades de convergencia de los siguients métodos para calcular $e$

* $\sum_{k=0}^{\infty} \frac{1}{k!}$
* $\lim_{n \to \infty} \left( 1 + \frac{1}{n} \right)^n$

```julia
#Construyo k! para despues poder usarla 
function fac(k)
    if k==1
        return 1
    elseif k>1
        return k*fac(k-1)
    else 
        print("No se puede calcular para")
        println(k)
        end 
end
```

```julia
#Construyendo la serie 
function eserie(n)
    suma=1 
    for k in 1:n
        suma=suma+(1/fac(k))
        end 
    return suma
end
```

```julia
#Construyendo la sucesión
function esucesion(n)
    l=(1+(1/n))^n
    return l
end 
```

Sabemos que para acercarnos al valor de e usamos las expresiones  
* $e=\sum_{k=0}^{\infty} \frac{1}{k!}$
* $e=\lim_{n \to \infty} \left( 1 + \frac{1}{n} \right)^n$

y que $e\approx 2,718 281 828 459 045 235 360...$
entonces graficando las expresiones anteriores tenemos

```julia
xs = 1:100
ys1 = [eserie(x) for x in xs]
ys2 = [esucesion(x) for x in xs]
scatter(xs,ys1,label="Serie",title="Comparación",alpha=0.7, xlabel="x", ylabel="y")
scatter!(xs,ys2,label="Sucesion",alpha=0.7)
```

se puede apreciar que la curva de serie converge más rapido al valor de $e$ que la de suceción, veamos en una gráfica log-log.

```julia
xs = 1:100
ys1 = [eserie(x) for x in xs]
ys2 = [esucesion(x) for x in xs]
# cambiar la escala a una gráfica log-log
scatter(xs,ys1,label="Serie",title="Comparación",alpha=0.7,xscale=:log10,yscale=:log10, xlabel="x",ylabel="y")
scatter!(xs,ys2,label="Sucesion",alpha=0.7)
```

Se sigue apreciando que la serie converge más rapido que la sucesión

Por otro lado, la función se puede aproximar mediante una serie de Fourier que toma la siguiente expresión

$$
g(x) =\frac{2}{\pi} \sum_{k=1}^n  (-1)^{k+1} \; \frac{\sin{(2 \pi k x)}}{k}
$$

Define una función `cuadrada(x,n)` que aproxime la onda de sierra con la suma mostrada. Grafica la función en el intervalo $[-1/2,1/2]$ para $n=5,8,11$, junto con la función $f$ original.


```julia
function sierra(x)
    f=2*x
    return f
end 
```

```julia
function cuadrada(x,n)
    suma=0
    for k in 1:n
        suma=suma+((-1)^(k+1))*(sin(2*π*k*x))/(k)
    end
    return suma*(2/π)
end
```
